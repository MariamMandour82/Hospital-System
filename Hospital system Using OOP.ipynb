```python
class Doctor:
    def __init__(self, name="", age="", specialty="", salary=""):
        self.name = name
        self.age = age
        self.specialty = specialty
        self.salary = salary

class Patient:
    def __init__(self, name="", age="", gender="", disease=""):
        self.name = name
        self.age = age
        self.gender = gender
        self.disease = disease

class Hospital:
    doctor_list = []
    patient_list = []

    def __init__(self):
        self.doctor_list = []
        self.patient_list = []

    def AddNewdoctor(self, doctor):
        self.doctor_list.append(doctor)

    def PrintAlldoctors(self):
        if len(self.doctor_list) == 0:
            print("No doctors At The Moment!")
        for doc in self.doctor_list:
            print(doc.name, "has age", doc.age, "has specialty", doc.specialty, "and salary", doc.salary)

    def DeleteByName(self):
        name_to_delete = input("Enter the name to delete: ")
        found = False
        for doc in self.doctor_list:
            if doc.name == name_to_delete:
                self.doctor_list.remove(doc)
                found = True
                break

        if found:
            print(f"doctor with the name '{name_to_delete}' has been deleted.")
        else:
            print(f"No doctor found with the name '{name_to_delete}'.")

    def UpdateSalaryByName(self):
        name = input("Enter the name ")
        new_salary = float(input("Enter the new salary: "))
        found = False
        for doc in self.doctor_list:
            if doc.name == name:
                doc.salary = new_salary
                found = True
                break
        if found:
            print("New salary has been updated")
        else:
            print("No doctor found with this name")

    def add_patient(self, patient):
        self.patient_list.append(patient)

    def display_patient_info(self):
        if len(self.patient_list) == 0:
            print("No Patients At The Moment!")
        for patient in self.patient_list:
            print(f"Name: {patient.name}, Age: {patient.age}, Gender: {patient.gender}, Medical Condition: {patient.disease}")


hospital = Hospital()

while True:
    print("\nProgram Options:")
    print("1) Doctor")
    print("2) Patient")
    print("3) End program")

    option = int(input("Enter your choice: "))

    if option == 1:
        while True:
            print("\nDoctor Program Options:")
            print("1) Add new doctor")
            print("2) Print all doctors")
            print("3) Delete doctor by name")
            print("4) Update doctor salary by name")
            print("5) End the program")

            doctor_option = int(input("Enter your choice: "))

            if doctor_option == 1:
                doctor = Doctor()
                doctor.name=input("Enter the name:")
                doctor.age=input("Enter age:")
                doctor.specialty=input("Enter specialty:")
                doctor.salary=input("Enter salary:")
                hospital.AddNewdoctor(doctor)
            elif doctor_option == 2:
                hospital.PrintAlldoctors()
            elif doctor_option == 3:
                hospital.DeleteByName()
            elif doctor_option == 4:
                hospital.UpdateSalaryByName()
            elif doctor_option == 5:
                print("Exiting doctor program.")
                break
            else:
                print("Invalid choice. Please try again.")

    elif option == 2:
        while True:
            print("\nPatient Program Options:")
            print("1) Add new patient")
            print("2) Display info of patients")
            print("3) End the program")

            patient_option = int(input("Enter your choice: "))

            if patient_option == 1:
                patient = Patient()
                patient.name=input("Enter the name:")
                patient.age=input("Enter age:")
                patient.gender=input("Enter gender:")
                patient.disease=input("Enter disease:")
                hospital.add_patient(patient)
            elif patient_option == 2:
                hospital.display_patient_info()
            elif patient_option == 3:
                print("Exiting patient program.")
                break
            else:
                print("Invalid choice. Please try again.")

    elif option == 3:
        print("Exiting the program.")
        break
    else:
        print("Invalid choice. Please try again.")
```
